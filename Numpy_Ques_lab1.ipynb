**Write a program to accept an array of elements and find the maximum without using max function**

```python
#Code here
array = [10, 5, 7, 20, 15, -9, 33]

if not array:
  print("array not provided")
else:
  max_num = array[0]
  for element in array[1:]:
    if element > max_num:
      max_num = element
  print(f"My array is {array}")
  print(f"Maximum number in array is {max_num}")
```

**Write a program to generate a ranodom matrix of size 5 X 6. Add another row containing random elements to it. Find the sum of each row.**

```python
#Code here
import numpy as np
rows = 5
cols = 6
lower_bound = 0
upper_bound = 9
matrix = np.random.randint(lower_bound, upper_bound, (rows, cols))
print(f"{matrix} \n")
new_random_row = np.random.randint(lower_bound, upper_bound, size=cols)
print(f"new added row is {new_random_row}")
matrix_with_random_row = np.vstack((matrix, new_random_row))
print(f"\n{matrix_with_random_row}")

row_sums = np.sum(matrix_with_random_row, axis=1)
print("\n sum of each row ")
print(row_sums)
```

**Write a program to generate a ranodm integer matrix of size 5 X 6. Add another column containing random integers to it. Compute the sine of each element and print it**

```python
#Code here
import numpy as np
rows = 5
cols = 6
lower_bound = 0
upper_bound = 9
matrix = np.random.randint(lower_bound, upper_bound, (rows, cols))
print(f"{matrix} \n")
new_random_col = np.random.randint(lower_bound, upper_bound, size=(rows,1))
print(f"new added column is {new_random_col}")
matrix_with_random_col = np.hstack((matrix, new_random_col))
print(f"\n{matrix_with_random_col}")

col_sums = np.sum(matrix_with_random_col, axis=0)
print("\n sum of each col ")
print(col_sums)

```
