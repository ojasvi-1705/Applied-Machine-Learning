```python
#Convert only numeric features into numpy array
feature_array = df.iloc[: , 0:4].to_numpy()
print(feature_array)


```

```python
# Find mean , median and standard deviation of each feature.
mean = np.mean(feature_array, axis=0)
median = np.median(feature_array, axis=0)
std_dev = np.std(feature_array, axis=0)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```

```python
#find the minimum and maximum petal length
min_petal_length = np.min(df['petal_length'])
print(min_petal_length)
max_petal_length = np.max(df['petal_length'])
print(max_petal_length)
```

```python
#find the corellation matrix of the faetures
correlation_matrix = np.corrcoef(feature_array, rowvar=False)
print(correlation_matrix)
```

```python
# add a new column/feature
#"petal area "which will be ( Petal length * petal width)

# Add new column
df["Petal Area"] = df.iloc[:, 2] * df.iloc[:, 3]
print(df.head())
```

```python
# plot histogram of sepal length
df.hist(column='sepal_length')
plt.show()
```

```python
#scatter plot of petal length vs petal width
df.plot(kind='scatter', x='petal_length',y = 'petal_width')
plt.show()
```

```python
#bar chart of average sepal length of each species
average_sepal_length = df.groupby("species")["sepal_length"].mean()

average_sepal_length.plot(kind='bar')
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")
plt.title("Average Sepal Length of Each Species")
plt.show()
```
